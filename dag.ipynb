```python
# Création du modèle de réseau bayésien (DAG)
# Création du modèle de réseau bayésien (DAG)
model = BayesianNetwork([('Publicité', 'Ventes'),
                         ('EcoImage', 'Ventes'),
                         ('TechImage', 'Ventes'),
                         ('Publicité', 'EcoImage'),
                         ('EcoImage', 'TechImage'),
                         ('R&D', 'TechImage')])  # Ajout de l'arc EcoImage -> TechImage

# Définir manuellement les positions des nœuds
pos = {
    'Publicité': (0, 2),
    'TechImage': (-1, 1),
    'EcoImage': (1, 1),
    'Ventes': (0, 0),
    'R&D' : (-1, 2)
}

# Dessiner le graphe avec les positions spécifiées
# Dessiner le graphe avec les positions spécifiées
nx.draw_networkx_nodes(model, pos, node_size=2000, node_color='lightblue')
nx.draw_networkx_edges(
    model, pos, arrowstyle='-|>', arrowsize=10, edge_color='black',
    #connectionstyle='arc3,rad=0.4',  # Ajustement du style de connexion
    min_source_margin=15, min_target_margin=20  # Ajustement des marges pour les flèches
)
nx.draw_networkx_labels(model, pos, font_size=7, font_weight='bold')

plt.show()
```
